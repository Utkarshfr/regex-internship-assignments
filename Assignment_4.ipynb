# Some basic explanations
- Longitude: represents how far west a house is
- Latitude: represents how far north a house is
- housing_median_age: represents median age of a house within a block
- total_bedrooms: Total number of rooms within a block
- population: population of a block
- households: represents how many families live within a block
- median_income: represents median income of households in tens of thousands of dollars
- median_house_value: represents median house value in a block
- ocean_proximity: represents how close a house is to  ocean (categorical data)

```python
# Plot between longitude and median_house_value
plt.figure(figsize = (12,12))
plt.style.use('seaborn-darkgrid')
sns.set(rc = {
    'axes.facecolor' : '#42f5bc',
    'axes.grid' : True,
    'xtick.labelsize' : 16,
    'ytick.labelsize' : 16
})
sns.lineplot(x = 'longitude',y = 'median_house_value',data = df,color = "red")
plt.title('longitude vs. house value')
```

# Inference
- The plot between longitude values and median_house_value does not give us a strong relation between the proximity of house towards west and its price. The hope was to get a general relation like if a house is more to the west we might find houses more expensive as compared to those which are located towards east of them or vice-versa.
- This plot is more of a random plot and does not suggest much of a strong relation between feature and median_house_value because their are price hikes at two spots between the graph at longitude = -122/-118 and rest points have low prices
- It can be inferred that longitude and house value does not generalize to some monotonous behaviour

```python
# Plot of latitude vs. House values
plt.figure(figsize = (12,12))
plt.style.use("seaborn-darkgrid")
sns.set(rc = {
    'axes.facecolor' : '#42f5bc',
    'axes.grid' : True,
    'xtick.labelsize' : 16,
    'ytick.labelsize' : 16
})
sns.lineplot(x = 'latitude',y = 'median_house_value',data = df,color = "red")
plt.title("latitude vs. house value")
```

# Inference
- Here as well we can not see much of a strong relation between latitude and the median house value because there are random spikes in the house values in the center of the graph and then again to the lower prices.

```python
# House value count of each ocean_proximity category
print(df['ocean_proximity'].value_counts())
```

```python
# Plot of House count of each ocean_proximity category across California
plt.figure(figsize = (15,6))
sns.set_style('whitegrid')
sns.countplot(y = 'ocean_proximity',data = df,palette = 'rainbow')
plt.title("House count across different regions of California")
```

```python
# Plot of median_house_value and house_median_age
plt.figure(figsize = (15,6))
sns.set_style('darkgrid')
sns.relplot(x = 'housing_median_age',y = 'median_house_value',palette = 'rainbow',data = df,kind = 'line')
plt.title("Cost of houses with respect to its age")
```

```python
# Histogram (frequency distribution) of houses based on median_house_value
plt.figure(figsize = (20,6))
sns.set_style('darkgrid')
sns.histplot(x = 'median_house_value',data = df, palette = 'rainbow',bins = 150)
plt.title("Median house value across California")
```

```python
# Distribution plot of houses based on median_house_value
plt.figure(figsize=(20,5))
sns.set_color_codes(palette="bright")
sns.distplot(df['median_house_value'],color='r',bins = 100)
plt.title("Median House Value across the state of California (CA)")
plt.ylabel("House count")
```

```python
# Average house value across California based on ocean_proximity category
plt.figure(figsize = (12,6))
sns.set_style('darkgrid')
sns.barplot(y = 'ocean_proximity',x = 'median_house_value',data = df,palette = 'rainbow')
plt.title("Average house value across California")
```

```python
# Distribution of houses by category over California
plt.figure(figsize = (8,8))
img = plt.imread('/content/california.png')

sns.scatterplot(x = 'longitude',y = 'latitude',data = df,hue = 'ocean_proximity')
plt.imshow(img,extent = [-124.55, -113.80, 32.45, 42.05])

plt.axis([-124.55, -113.80, 32.45, 42.05])
plt.title("House distribution across California state")
```
